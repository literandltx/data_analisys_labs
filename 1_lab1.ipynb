```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('resources/google_review_ratings.csv')

#  config dataframe
pd.set_option('display.expand_frame_repr', True) 
pd.set_option('display.max_columns', None) 

# before
df.info()
```

```python
columns_to_drop = [
    "User", 
    "Local services",
    "Unnamed: 25"
]

df = df.drop(columns=columns_to_drop)

df.info()
# print(df.head(5))
```

```python
# delete null or na
# print(df.isnull().mean())
df = df.dropna()
# print(df.isnull().mean())
```

```python
corr = df.corr()
fig, ax = plt.subplots(figsize=(15, 15))
sns.heatmap(corr,  cmap='coolwarm', annot=True, fmt=".1f")
plt.show()
```

```python
feature_names = df.columns
target = "Parks"

# for i in range(len(feature_names)):
for i in range(2):
    figure = plt.figure()
    ax = sns.boxplot(x=target, y=feature_names[i], data=df)

print("finished")
```
